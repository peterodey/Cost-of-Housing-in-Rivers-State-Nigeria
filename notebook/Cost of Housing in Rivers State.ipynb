# Definition of the Problem

There has been a growing increase in rural-urban migration in Rivers State, especially in Port Harcourt the capital city. This and other factors including industrialization, social stratification, and the global economic inflation significantly affects the cost variation of housing across Port Harcourt city. Hense, it is pertinent to investigate the cost of housing in Port Harcourt and other Local Government Areas in Rivers State.

# Project Aim and Objectives

This project aimed at understanding the cost of housing in Rivers State of Nigeria.
The objectives of this project include:
- To Investigate the Cost of Different Apartments in Selected LGAs in Rivers State
- To Investigate the Cost of Different Apartments in Selected Major Towns in Port Harcourt
- To Investigate the Average Cost of Apartments in Selected Major Towns Within Port Harcourt City

# Data Source and Methodology

The data was obtained from residents of Rivers State through the use of questionnaires (Google forms). It was cleaned using Power BI, then visualized using the python libraries; pandas, seaborn and matplotlib.

# Ethical Considerations

We obtaining informed consent from participant, ensuring their anonymity, confidentiality and voluntary participation. We adhere to legal and ethical standards, respected cultural differences, maintaining transparency and accountability throughout the data collection process.

# Data Analysis/Visualization

```python
# Importing the necessary python libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
df = pd.read_csv('Cleaned Rivers Housing Data-1.csv')
```

```python
# Display the dataset
df
```

```python
# Checking for a summary of the dataframe
df.info()
```

```python
# Descriptive Statistics of the dataset
df.describe()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 10))

# Pie Chart for Different Types of Accommodation Type
df['Accommodation Type'].value_counts().plot(kind='pie', autopct='%1.0f%%', startangle=140, ax=axes[0])
axes[0].set_title('Distribution of Apartments by Accommodation Type')
axes[0].set_ylabel('')

# Pie Chart for Different Types of Apartment
df['Type of Apartment'].value_counts().plot(kind='pie', autopct='%1.0f%%', startangle=140, ax=axes[1])
axes[1].set_title('Distribution of Apartments by Type of Apartment')
axes[1].set_ylabel('')

plt.tight_layout()
plt.savefig('artifact/Different_Types_of_Accommodation.pdf', format='pdf')
plt.savefig('image/Different_Types_of_Accommodation.jpg')
plt.show()
```

```python
# Calculating average costs
avg_costs = df.groupby('LGA')['Cost of Apartment'].mean().sort_values(ascending=False)

# Setting up figure and axis
plt.figure(figsize=(12, 8))

# Plot bar chart with custom color palette
plt.bar(avg_costs.index, avg_costs.values, color='skyblue')

# Adding labels and title
plt.xlabel('LGAs')
plt.ylabel('Average Cost of Apartments')
plt.title('Average Cost of Apartments Across LGAs')

# Rotating x-axis labels for better readability
plt.xticks(rotation=60)

# Save picture as pdf and jpg
plt.savefig('artifact/Average_Cost_of_Apartments_Across_LGAs.pdf', format='pdf')
plt.savefig('image/Average_Cost_of_Apartments_Across_LGAs.jpg')

# Show plot
plt.show()
```

# To Investigate the Cost of Different Apartments in Selected LGAs in Rivers State

```python
# Define features of the categories
accommodation_categories = ['One Room', 'Two Rooms', 'Self Contain', 'One Bedroom Flat', 'Two Bedroom Flat', 'Three Bedroom Flat', 
                            'Four Bedroom Flat', 'Five Bedroom Flat', 'One Bedroom Duplex', 'Two Bedroom Duplex', 
                            'Three Bedroom Duplex', 'Four Bedroom Duplex', 'Five Bedroom Duplex']

# Create profiles based on features of the categories
plt.figure(figsize=(15, 30))

for i, feature in enumerate(accommodation_categories, 1):
    plt.subplot(7, 2, i)
    avg_costs_by_LGA = df.groupby('LGA')[feature].mean().sort_values(ascending=False)
    sns.barplot(x=avg_costs_by_LGA.index, y=avg_costs_by_LGA.values, palette='viridis')
    plt.title(f'Average Cost of Accommodation Across LGAs For {feature.capitalize()}')
    plt.xlabel('LGAs')
    plt.ylabel('Average Cost of Apartment')
    plt.xticks(rotation=45)

plt.tight_layout()

# Save picture as pdf and jpg
plt.savefig('artifact/Average_Cost_of_Accommodation_Across_LGAs.pdf', format='pdf')
plt.savefig('image/Average_Cost_of_Accommodation_Across_LGAs.jpg')
plt.show()
```

# To Investigate the Cost of Different Apartments in Selected Major Towns in Port Harcourt 

```python
# Define features of the categories
accommodation_categories = ['One Room', 'Two Rooms', 'Self Contain', 'One Bedroom Flat', 'Two Bedroom Flat', 'Three Bedroom Flat', 
                            'Four Bedroom Flat', 'Five Bedroom Flat', 'One Bedroom Duplex', 'Two Bedroom Duplex', 
                            'Three Bedroom Duplex', 'Four Bedroom Duplex', 'Five Bedroom Duplex']

# Create profiles based on features of the categories
plt.figure(figsize=(15, 30))

for i, feature in enumerate(accommodation_categories, 1):
    plt.subplot(7, 2, i)
    avg_costs_by_Town = df.groupby('Town')[feature].mean().sort_values(ascending=False)
    sns.barplot(x=avg_costs_by_Town.index, y=avg_costs_by_Town.values, palette='viridis')
    plt.title(f'Average Cot of Accommodation Across Towns For {feature.capitalize()}')
    plt.xlabel('Towns')
    plt.ylabel('Average Cost of Apartments')
    plt.xticks(rotation=65)

plt.tight_layout()

# Save picture as pdf and jpg
plt.savefig('artifact/Average_Cost_of_Accommodation_Across_Towns.pdf', format='pdf')
plt.savefig('image/Average_Cost_of_Accommodation_Across_Towns.jpg')
plt.show()
```

# To Investigate the Average Cost of Apartments in Selected Major Towns Within Port Harcourt City

```python
# Define accommodation categories
accommodation_categories = ['One Room', 'Two Rooms', 'Self Contain', 'One Bedroom Flat', 'Two Bedroom Flat',
                            'Three Bedroom Flat', 'Four Bedroom Flat', 'Five Bedroom Flat', 'One Bedroom Duplex',
                            'Two Bedroom Duplex', 'Three Bedroom Duplex', 'Four Bedroom Duplex', 'Five Bedroom Duplex']

# Define towns of interest
towns_of_interest = ['Rumuomasi', 'Rumuokwurusi', 'Rumuagholu', 'Choba', 'Rukpoku', 'Rumuosi', 'Elelenwo', 'Alakahia',
                     'Ozuoba', 'Mgbuoba', 'Rumudara', 'Eliozu', 'Nkpolu', 'Agip', 'Rumuepirikom', 'Ada George',
                     'Rumuigbo', 'Oroazi', 'D-Line', 'Woji', 'Mile 4', 'Rumunduru']

# Create profiles based on features of the categories
plt.figure(figsize=(15, 40))  # Adjust the figure size

for i, category in enumerate(accommodation_categories, 1):
    plt.subplot(12, 2, i)
    sns.barplot(x='Town', y=category, data=df[df['Town'].isin(towns_of_interest)], palette='viridis',
                order=df[df['Town'].isin(towns_of_interest)].groupby('Town')[category].mean().sort_values(ascending=False).index,
                errorbar=None,
                linewidth=0)
    plt.title(f'Average Cost of {category} Across Towns Within Port Harcourt City')
    plt.xlabel('Towns')
    plt.ylabel('Average Cost')
    plt.xticks(rotation=65)
    plt.tight_layout()

# Save picture as pdf and jpg
plt.savefig('artifact/Average_Cost_of_Accommodation_Within_Port_Harcourt.pdf', format='pdf')
plt.savefig('image/Average_Cost_of_Accommodation_Within_Port_Harcourt.jpg')
plt.show()
```

## Findings/Conclusion
- LGAs with most expensive houses are Port Harcourt City and Obio/Akpor, while those with least expensive houses are Khana, Emuoha and Obigbo.
- Towns with most expensive houses are Rumuepirekom, Mgbuoba, and Woji, while those with least expensive houses are Bori, Emuoha, Obigbo, and Mile 4.
- Towns with most expensive houses within Port Harcourt City are Rumuepirekom and Mgbuoba, while those with least expensive houses are Mile 4 and Rumuosi.
